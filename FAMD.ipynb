## FAMD

```python
import prince
```

```python
famd = prince.FAMD(
    n_components=5,
    n_iter=3,
    copy=True,
    check_input=True,
    random_state=42,
    engine="sklearn",
    handle_unknown="error"  # same parameter as sklearn.preprocessing.OneHotEncoder
)
famd = famd.fit(df)
```

n_components — the number of principal components to computed. You only need two if your intention is to visualize the two major components.주성분 수

n_iter — the number of iterations used for computing the SVD. 특이값분해시 반복 수

copy — Whether nor to perform the computations inplace. if False then the computations will be done inplace which can have possible side-effects on the input data

check_input — Whether to check the coherence of the inputs or not.

engine — what SVD engine to use (should be one of ['fbpca', 'sklearn'])

random_state — controls the randomness of the SVD results.

```python
from prince import FAMD

famd = FAMD(n_components=10)  # 원하는 주성분 개수 설정
famd.fit(df)  # 훈련 데이터에 FAMD 적용

# 변환 매개변수 저장
famd_params = {
    'means': famd.row_coordinates(df).mean(axis=0),
    'stds': famd.row_coordinates(df).std(axis=0),
    'components': famd.column_correlations(df)
}
```

```python
famd_params
```

```python
famd.eigenvalues_summary
```

```python
famd.row_coordinates(df).head()
```

```python
famd.column_coordinates_
```

```python
(
    famd.row_contributions_
    .sort_values(0, ascending=False)
    .head(5)
    .style.format('{:.3%}')
)
```

```python
famd.column_contributions_.style.format('{:.0%}')
```
