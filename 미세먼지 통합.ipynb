```python
import pandas as pd
import numpy as np
```

```python
n = [0]*24
df = []
for i in range(1,25):
    n[i-1] = pd.read_excel(f'미세먼지/서울{i}.xls',header=None)
    duplicated_df = pd.DataFrame(np.repeat(n[i-1].iloc[2:].values, 6, axis=0)) 
    df.append(duplicated_df)
```

```python
merged_df = pd.concat(df)
merged_df
```

```python
merged_df = merged_df.iloc[:, [1, 2]] 
merged_df.columns = ['pm10', 'pm2.5']
```
