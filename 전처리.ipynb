```python
import pandas as pd
import numpy as np
import seaborn as sns  
import matplotlib.pyplot as plt
```

```python
month = ["01","02","03","04","05","06","07","08","09","10","11","12"]
df_2020 = pd.DataFrame()

for i in range(0,12):
    _month = month[i]
    df = pd.read_csv('2020/2020'+_month+'_uv.csv')
    df.drop(columns='Unnamed: 0', axis=1, inplace=True)
    df.columns = ['Date','Time','stn','lon','lat','uv','band1','band2','band3','band4','band5','band6','band7','band8','band9',\
                  'band10','band11','band12','band13','band14','band15','band16','solarza','sateza','esr','height','landtype']
    df_2020 = pd.concat([df_2020, df])
```

```python
df_2021 = pd.DataFrame()

for i in range(0,12):
    _month = month[i]
    df = pd.read_csv('2021/2021'+_month+'_uv.csv')
    df.drop(columns='Unnamed: 0', axis=1, inplace=True)
    df.columns = ['Date','Time','stn','lon','lat','uv','band1','band2','band3','band4','band5','band6','band7','band8','band9',\
                  'band10','band11','band12','band13','band14','band15','band16','solarza','sateza','esr','height','landtype']
    df_2021 = pd.concat([df_2021, df])
```

```python
df_all = pd.DataFrame()
df_all = pd.concat([df_2020, df_2021])
```

```python
df_all = df_all.astype({'Time':str})
for i in range(len(df_all)):
    if df_all.iloc[i,1] == '0':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '10':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '20':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '30':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '40':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '110':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '120':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '130':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '140':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '210':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '220':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '230':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
    elif df_all.iloc[i,1] == '240':
        df_all.iloc[i,1] = df_all.iloc[i,1].zfill(4)
```

```python
# datetime속성의 새로운 칼럼 생성
df_all['Datetime'] = pd.to_datetime(df_all['Date'].astype(str) + df_all['Time'].astype(str), format='%Y%m%d%H%M')
df_all
```

```python
stn_index = df_all['stn'].value_counts().index
df_stn = [0]*15
for i in range(15):
    df_stn[i] = df_all.groupby('stn').get_group(stn_index[i])
```
