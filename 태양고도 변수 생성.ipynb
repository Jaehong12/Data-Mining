태양천정각에 이상치가 존재함
(동일한 시간대에 이상치 존재)

### 태양천정각 이상치 처리

```python
"""
계절성과 주기를 고려한 이상치를 null 값으로 대체하는 함수

Parameters:
data (pd.Series): 대상 시계열 데이터
window_size (int): 이동평균 적용 윈도우 크기
threshold (float): 이상치로 간주하는 값의 기준치

Returns:
pd.Series: 이상치가 null 값으로 대체된 시계열 데이터
"""
def replace_outliers_seasonal(data, window_size=12, threshold=2):
    # 이동평균 계산
    rolling_mean = data.rolling(window=window_size, center=True).mean()

    # 계절성 추출
    seasonal = data - rolling_mean

    # 이상치 기준치 계산
    std = np.std(seasonal)
    cutoff = std * threshold

    # 이상치를 null 값으로 대체
    seasonal[(seasonal < -cutoff) | (seasonal > cutoff)] = np.nan

    # 이동평균으로 대체
    cleaned = seasonal + rolling_mean

    return cleaned
```

```python
for i in range(15):
    stn[i].loc[12:,'solarza'] = replace_outliers_seasonal(stn[i].loc[12:,'solarza'])
```

```python
for i in range(15):
    stn[i]['solarza'] = stn[i]['solarza'].interpolate(method = 'linear')
```
